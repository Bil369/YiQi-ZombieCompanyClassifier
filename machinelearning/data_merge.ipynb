```python
import pandas as pd
```

```python
base_train_sum = pd.read_csv('datasets/base_train_sum.csv', encoding='gbk')
```

```python
base_train_sum['ID'] = base_train_sum['ID'].astype('int')
```

```python
base_train_sum
```

```python
base_verify = pd.read_csv('datasets/base_verify1.csv', encoding='gbk')
```

```python
base_verify['ID'] = base_verify['ID'].astype('int')
```

```python
base_verify.drop('控制人ID', axis=1, inplace=True)
```

```python
base_verify
```

```python
base_total_sum = pd.concat([base_train_sum, base_verify]).reset_index().drop('index', axis=1)
```

```python
base_total_sum
```

```python
base_total_sum.to_csv('datasets/base_total_sum.csv', encoding='gbk', index=False)
```
